```python
# Python menggunakan logika boolean untuk mengevaluasi kondisi (true/false)
print('⬇⬇⬇ mengevaluasi kondisi "True" atau "False" ⬇⬇⬇')
kondisi = 2
k1 = kondisi == 2
k2 = kondisi == 3
k3 = kondisi < 5
print("Apakah kondisi sama dengan 2? %s" % k1)
print("Apakah kondisi sama dengan 3? %s" % k2)
print("Apakah kondisi dibawah 5? %s" % k3)
# Masih banyak operator lainnya seperti :
# != (tidak sama dengan)
# <= (lebih kecil atau sama dengan)
# >= (lebih besar atau sama dengan)
# < (lebih kecil dari)
# > (lebih besar dari)

# ==============================================================================================
print("=========================================================================================")
# Operator boolean "and" dan "or" memungkinkan pembuatan ekspresi boolean menjadi lebih kompleks
print('⬇⬇⬇ menggunakan operator "and" dan "or" untuk mengecek kondisi ⬇⬇⬇')
namanya = "galih"
absen = 17
if namanya == "galih" and absen == 17 :
    print("Nama kamu adalah %s, nomor absen %d." % (namanya, absen))
if namanya == "galih" or namanya == "sukma" :
    print("Nama kamu adalah %s." % namanya)

# ==============================================================================================
print("=========================================================================================")
# Operator "in" digunakan untuk memeriksa apakah objek tertentu ada di dalam wadah objek, seperti daftar
print('⬇⬇⬇ operator "in" untuk mengecek isi object ⬇⬇⬇')
namaKetua = "adjie"
if namaKetua in ["galih", "sukma", "adjie"] :
    print("Nama anggota kelompok ada galih, sukma, dan adjie, ketuanya adalah %s." % namaKetua)

# ==============================================================================================
print("=========================================================================================")
# Pembuatan blok kode ekspresi
print("⬇⬇⬇ membuat blok dengan kode expresi ⬇⬇⬇")
pernyataan = False
pernyataanLain = True

if pernyataan is True :
    print("Pernyatan : %s." % pernyataan)
    pass
elif pernyataanLain is True :
    print("Pernyataan lain : %s." % pernyataanLain)
    pass
else :
    print("Tidak ada pernyataan yang sesuai")
    pass

# ==============================================================================================
print("=========================================================================================")
# Operator "==" tidak sama dengan "is" 
# Operator "is" tidak cocok dengan nilai variabel, tetapi instance itu sendiri
print('⬇⬇⬇ perbedaan operator "is" dan "==" ⬇⬇⬇')
xVar = [1,2,3]
yVar = [1,2,3]
print(xVar == yVar) # True
print(xVar is yVar) # False

# ==============================================================================================
print("=========================================================================================")
# Menggunakan "not" sebelum boolean akan meng-invert nilainya
print('⬇⬇⬇ invert nilai boolean dengan "not" ⬇⬇⬇')
print(not False)
print((not False) == (False) )

# ==============================================================================================
print("=========================================================================================")
# Kesimpulan dan latihan
print("⬇⬇⬇ kesimpulan dan latihan ⬇⬇⬇")
# Ubah variabel, sehingga semua pernyataan if diselsesaikan sebagai TRUE
angka = 17
angka_kedua = 0
array_pertama = [3,1,4]
array_kedua = [2,1]

# Soal nomor 1
sp1 = 15
if angka > sp1 :
    print("Soal nomor 1 hasilnya :", angka > sp1)
else :
    print("Soal nomor 1 hasilnya :", angka > sp1)

# Soal nomor 2
if array_pertama :
    print("Soal nomor 2 hasilnya :", not not array_pertama)
else :
    print("Soal nomor 2 hasilnya :", not not array_pertama)

# Soal nomor 3
sp2 = 2
if len(array_kedua) == sp2 :
    print("Soal nomor 3 hasilnya :", len(array_kedua) == sp2)
else :
    print("Soal nomor 3 hasilnya :", len(array_kedua) == sp2)

# Soal nomor 4
sp3 = 5
if len(array_pertama) + len(array_kedua) == sp3 :
    print("Soal nomor 4 hasilnya :", len(array_pertama) + len(array_kedua) == sp3)
else :
    print("Soal nomor 4 hasilnya :", len(array_pertama) + len(array_kedua) == sp3)

# Soal nomor 5
if array_pertama and array_pertama[1] == 1 :
    print("Soal nomor 5 hasilnya :", array_pertama and array_pertama[1] == 1)
else :
    print("Soal nomor 5 hasilnya :", array_pertama and array_pertama[1] == 1)

# Soal nomor 6
if not angka_kedua :
    print("Soal nomor 6 hasilnya :", not angka_kedua)
else :
    print("Soal nomor 6 hasilnya :", not angka_kedua)
```

# **Functions**
#### **Apa itu fungsi?**
Fungsi adalah cara mudah untuk membagi kode Anda menjadi blok-blok yang berguna, memungkinkan kita mengurutkan kode kita, membuatnya lebih mudah dibaca, menggunakannya kembali, dan menghemat waktu. Juga fungsi adalah cara utama untuk mendefinisikan antarmuka sehingga pemrogram dapat membagikan kode mereka.
#### **Bagaimana menulis fungsi dengan python?**
Seperti pada materi sebelumnya, Python menggunakan blok. Di mana garis blok lebih banyak kode Python, dan kepala blok memiliki format berikut: 
<p>kepala_blok :</p>
<ul>
    <li>blok line pertama</li>
    <li>blok line kedua</li>
    <li>...</li>
</ul>
block_keyword block_name(argument1,argument2, ...) Kata kunci blok yang sudah Anda ketahui adalah "if", "for", dan "while".

Fungsi dalam python didefinisikan menggunakan kata kunci blok "def", diikuti dengan nama fungsi sebagai nama blok.

#### **Bagaimana Anda memanggil fungsi dengan Python?**
Cukup tulis nama fungsi diikuti dengan (), tempatkan argumen yang diperlukan di dalam tanda kurung. Contoh:

```python
# penggunaan fungsi sederhana dengan isi blok print()
def fungsi_pertama() :
    print("⬇⬇⬇ Fungsi Pertama ⬇⬇⬇")
    print("Contoh penggunaan fungsi sederhana")
    print("=========================================================================================")
# ==============================================================================================

# fungsi dapat menerima argumen (variabel diteruskan dari pemanggil ke fungsi)
def fungsi_kedua(nama, kelas) :
    print("⬇⬇⬇ Fungsi kedua ⬇⬇⬇")
    print("Halo, Nama saya %s, saya kelas %s." % (nama,kelas))
    print("=========================================================================================")
# ==============================================================================================

# Fungsi dapat mengembalikan nilai ke pemanggil, menggunakan kata kunci- 'return'
def fungsi_ketiga(a,b) :
    return a + b            # mengembalikan hasil perhitungan nilai a + b
# ==============================================================================================

# ⬇⬇⬇ Penggunaan Fungsi ⬇⬇⬇
# memanggil fungsi pertama
fungsi_pertama()

# memanggil fungsi kedua
fungsi_kedua("Galih", "TRKE-7A")

# memanggil fungsi ketiga
x = fungsi_ketiga(2, 3)     # nilai a dan b diambil dari sini dan hasilnya dimasukkan ke variabel "x"
print("⬇⬇⬇ Fungsi ketiga ⬇⬇⬇")
print(x)

# ==============================================================================================
print("=========================================================================================")
# kesimpulan dan latihan
print("⬇⬇⬇ Kesimpulan dan Latihan ⬇⬇⬇")
# Dalam latihan ini Anda akan menggunakan fungsi yang sudah ada, dan sambil menambahkan fungsi 
# Anda sendiri untuk membuat program yang berfungsi penuh.
# 1. Tambahkan fungsi bernama list_keuntungan() yang mengembalikan daftar string berikut:
#    "semua tugas jadi mudah", "mengetahui banyak hal", "lebih percaya diri", 
#    "dapat membantu banyak orang dengan ilmu kita"
# 2. Tambahkan fungsi bernama buat_kalimat(info) yang menerima satu argumen berisi string dan mengembalikan 
#    kalimat yang dimulai dengan string yang diberikan dan diakhiri dengan string "adalah manfaat belajar"
# 3. Jalankan dan lihat semua fungsi bekerja sama!


# fungsi pertama memuat keuntungan
def list_keuntungan() :
    return "semua tugas jadi mudah", "mengetahui banyak hal", "lebih percaya diri", "dapat membantu banyak orang dengan ilmu kita"

# fungsi kedua memuat sumber keuntungan
def buat_kalimat(benefit) :
    return "%s adalah manfaat belajar" % benefit

# fungsi ketiga menggabungkan 2 fungsi diatas
def sebutkan_manfaat_belajar() :
    list_dari_keuntungan = list_keuntungan()
    for benefit in list_dari_keuntungan :
        print(buat_kalimat(benefit))
# menjalankan fungsi terakhir dimana semua fungsi telah digabungkan
sebutkan_manfaat_belajar()
```

# **Classes and Objects**
Objek adalah enkapsulasi variabel dan fungsi menjadi satu kesatuan. Objek mendapatkan variabel dan fungsinya dari classes. Classes pada dasarnya adalah template untuk membuat objek Anda.

```python
# Dasar penggunaan class seperti dibawah ini dapat dilihat dibawah ini
class contohClass :
    variabel = "contoh variabel pada class"

    def fungsi(dirimu) :
        print("Ini adalah pesan yang ada didalam class")
# Nanti akan ada penjelasan mengapa diharuskan menyertakan parameter "dirimmu". 
# Pertama, untuk menetapkan class (template) di atas ke objek, Anda akan melakukan hal berikut:

class contohClass_1 :
    variabel_1 = "contoh variabel_1 pada class"

    def fungsi_1(dirimu_1) :
        print("Ini adalah pesan yang ada didalam class_1 dan fungsi_1")

objekSaya = contohClass_1()
# Sekarang variabel "objekSaya" menyimpan objek kelas "contohClass_1" yang berisi variabel dan fungsi yang 
# didefinisikan di dalam kelas yang disebut "contohClass_1".

# ==============================================================================================
# ⬇⬇⬇ MENGAKSES VARIABEL OBJEK ⬇⬇⬇
# untuk dapat mengakses variabel di dalam objek "objekSaya_1", dapat dilakukan dengan cara berikut :
print('⬇⬇⬇ Mengakses Variabel di dalam objek "objekSaya_1" ⬇⬇⬇')
class contohClass_2 :
    variabel_2 = "contoh variabel_2 pada class"

    def fungsi_2(dirimu_2) :
        print("Ini adalah pesan yang ada didalam class_2 dan fungsi_2")

objekSaya_1 = contohClass_2()

objekSaya_1.variabel_2
# atau dapat ditampilkan sebagai string seperti dibawah ini
print(objekSaya_1.variabel_2)

# ==============================================================================================
print("=========================================================================================")
# Anda dapat membuat beberapa objek berbeda yang berasal dari kelas yang sama (memiliki variabel dan fungsi yang sama yang ditentukan). 
# Namun, setiap objek berisi salinan independen dari variabel yang didefinisikan dalam kelas. 
# Misalnya, jika kita mendefinisikan objek lain dengan kelas "contohClass_3" dan kemudian mengubah string dalam variabel di atas:
print("⬇⬇⬇ Membuat beberapa objek dari kelas dan variabel yang sama ⬇⬇⬇")
class contohClass_3 :
    variabel_3 = "isi variabel_3 default"

    def fungsi_3(dirimu_3) :
        print("Ini adalah pesan yang ada didalam class_3 dan fungsi_3")

objekSaya_2x = contohClass_3()
objekSaya_2y = contohClass_3()

objekSaya_2y.variabel_3 = "isi variabel_3 dirubah pada objek ini"
# kemudian cetak masing masing nilai pada objekSaya
print(objekSaya_2x.variabel_3)
print(objekSaya_2y.variabel_3)

# ==============================================================================================
print("=========================================================================================")
# Untuk mengakses fungsi di dalam objek, Anda menggunakan notasi yang mirip dengan mengakses variabel:
print("⬇⬇⬇ Mengakses fungsi objek ⬇⬇⬇")
class contohClass_4 :
    variabel_4 = "contoh variabel_4 pada class"

    def fungsi_4(dirimu_4) :
        print("Ini adalah pesan yang ada didalam class_4 dan fungsi_4")

objekSaya_3 = contohClass_4()

objekSaya_3.fungsi_4() # memanggil fungsi_4 yang ada didalam contohClass_4

# ==============================================================================================
print("=========================================================================================")
# Fungsi __init__(), adalah fungsi khusus yang dipanggil saat kelas sedang diinisiasi. Ini digunakan untuk menetapkan nilai di kelas.
print('⬇⬇⬇ Fungsi "__init__()" ⬇⬇⬇')
class NumberHolder :
    def __init__(self, nomor) :
        self.angka = nomor

    def returnNumber(self):
        return self.angka
    
var = NumberHolder(3)
print(var.returnNumber())

# ==============================================================================================
print("=========================================================================================")
# kesimpulan dan latihan
print("⬇⬇⬇ Kesimpulan dan Latihan ⬇⬇⬇")
# Kami memiliki kelas yang ditentukan untuk kendaraan. Buat dua kendaraan baru bernama "mobil1" dan "mobil2".
# Tetapkan "mobil1" menjadi mobil Hypercar biru senilai $700.000,00 dengan nama Galih, 
# dan "mobil2" menjadi Rangerover putih bernama Sukma senilai $300.000,00.

# mendefinisikan class kendaraan
class kendaraan :
    nama = ""
    jenis = "mobil"
    warna = ""
    nilai = 100.00
    def deskripsi(sendiri) :
        deskripsi_mobil = "%s membeli %s berwarna %s seharga $%.2f." % (sendiri.nama, sendiri.jenis, sendiri.warna, sendiri.nilai)
        return deskripsi_mobil
    
# tulis kode kalian dibawah ini
mobil1 = kendaraan()
mobil1.nama = "Galih"
mobil1.jenis = "Hypercars"
mobil1.warna = "biru"
mobil1.nilai = 700000.00

mobil2 = kendaraan()
mobil2.nama = "Sukma"
mobil2.jenis = "Rangerover"
mobil2.warna = "putih"
mobil2.nilai = 300000.00

# kode untuk pengecekan hasil latihan
print(mobil1.deskripsi())
print(mobil2.deskripsi())
```

# **Dictionaries**

Dictionaries adalah tipe data yang mirip dengan array, tetapi bekerja dengan kunci dan nilai, bukan indeks.
Setiap nilai yang disimpan dalam Dictionaries dapat diakses menggunakan kunci, yang merupakan jenis objek apa pun (string, angka, daftar, dll.) alih-alih menggunakan indeksnya untuk mengatasinya.

```python
print('⬇⬇⬇ Mencetak isi bukuTelfon dengan "Dictionaries" ⬇⬇⬇')
# Contoh, database nomor telepon dapat disimpan menggunakan "Dictionaries"
bukuTelfon = {}
bukuTelfon["Galih"] = 8990774858
bukuTelfon["Sukma"] = 8775538833
bukuTelfon["adjie"] = 8996677332

print(bukuTelfon)

# ==============================================================================================
print("=========================================================================================")
# Notasi "Dictionaries" diatas dapat juga dituliskan dengan cara lain seperti dibawah ini
print('⬇⬇⬇ Notasi alternatif "Dictionaries" ⬇⬇⬇')
daftarKontak = {
    "Galih": 8990774858,
    "Sukma": 8775538833,
    "adjie": 8996677332
}

print(daftarKontak)

# ==============================================================================================
print("=========================================================================================")
# "Dictionaries" dapat diulang, seperti "List".
# Namun, "Dictionaries", tidak seperti "List", tidak menyimpan urutan nilai yang disimpan di dalamnya.
print('⬇⬇⬇ Mengulangi "Dictionaries" ⬇⬇⬇')
kontakTim = {
    "Ketua": 8990774858,
    "Wakil": 8775538833,
    "Admin": 8996677332,
    "Bendahara": 8995522552,
    "Anggota": 8966325745
}
for nama, nomor in kontakTim.items():
    print("Nomor telfon dari %s tim kami adalah %d" % (nama, nomor))

# ==============================================================================================
print("=========================================================================================")
# Untuk menghapus indeks tertentu, gunakan salah satu dari notasi dibawah ini
print('⬇⬇⬇ Menghapus nilai dari "Dictionaries" ⬇⬇⬇')
print("Notasi 1 ⬇")
deleteKontak = {
    "Galih": 8990774858,
    "Sukma": 8775538833,
    "Adjie": 8996677332
}
del deleteKontak["Sukma"]
print(deleteKontak)
# Atau dengan notasi lainnya
print("Notasi 2 ⬇")
hapusKontak = {
    "Galih": 8990774858,
    "Sukma": 8775538833,
    "Adjie": 8996677332
}
hapusKontak.pop("Adjie")
print(hapusKontak)

# ==============================================================================================
print("=========================================================================================")
# kesimpulan dan latihan
print("⬇⬇⬇ Kesimpulan dan Latihan ⬇⬇⬇")
# Tambahkan "Sukma" ke "kontakWA" dengan nomor telepon 938273443, dan hapus "Jill" dari "kontakWA".
kontakWA = {
    "galih": 8996677332,
    "Udin": 8996677332,
    "Adjie": 8996677332
}

kontakWA["Sukma"] = 938273443
del kontakWA["Udin"]

# Kode untuk pengecekan hasil latihan
if "Sukma" in kontakWA :
    print("Sukma telah terdaftar di kontakWA")
if "Udin" not in kontakWA :
    print("Udin di kick dari kontakWA")
```

# **Modules and Packages**

Dalam pemrograman, modul adalah perangkat lunak yang memiliki fungsi tertentu.
Misalnya, saat membuat game ping pong, satu modul mungkin bertanggung jawab atas logika game, dan modul lain menggambar permainan di layar.
Setiap modul terdiri dari file yang berbeda, yang dapat diedit secara terpisah.

#### **Menulis modul**
Modul dalam Python hanyalah file Python dengan ekstensi .py. Nama modul sama dengan nama file.
Modul Python dapat memiliki sekumpulan fungsi, kelas, atau variabel yang didefinisikan dan diimplementasikan.
Contoh di dibawah menyertakan dua file modul:

videoGame/
<ul>
    <li>videoGame/game.py</li>
    <li>videoGame/draw.py</li>
</ul>

Script python "game.py" mengimplementasikan permainan. ia menggunakan modul draw dari script "draw.py" untuk menggambar game di layar.
Modul diimpor dari modul lain menggunakan perintah "import".

```python
bahasa = "Python"
level = "mudah"
doa = "Belajar %s itu %s" % (bahasa, level)
print(doa)
```
